```python
# Customized colors
bg = '#F9F9F9'
```

## Distributions

The cell below calculate and plot pdfs of some distributions often used in econometrics.

```python
# Define parameters and a probability model.
μ, σ = 0, 1
std_normal = stats.norm(μ, σ)

# Get the support.
lb, ub = std_normal.support()

# Calculate the first four moments.
mean, var, skew, kurt = std_normal.stats(moments='mvsk')

# Display the probability density function.
x = np.linspace(std_normal.ppf(0.001),
                std_normal.ppf(0.999),
                100)
q = std_normal.ppf(0.95)

fig, ax = plt.subplots()

ax.plot(x, std_normal.pdf(x), color='dodgerblue', label='Std. Normal PDF')
ax.fill_between(x[x < q], 0, std_normal.pdf(x[x < q]), color='dodgerblue', alpha=0.3)

ax.set_title('Probability Distribution Function of $N(0, 1)$')
ax.set_facecolor(bg)
ax.set_xmargin(0)
ax.set_ylim([0, None])
```
